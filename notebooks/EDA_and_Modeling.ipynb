```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load raw dataset
df = pd.read_csv(r"C:\Guvi\NutriClass Food Classification Using Nutritional Data\data\raw\synthetic_food_dataset_imbalanced.csv")

# Display basic info
print(df.info())
print(df.describe())
print(df.head())

# Check target distribution
target_col = 'Meal_Type'
print(df[target_col].value_counts())

# Visualize class distribution
plt.figure(figsize=(8,6))
sns.countplot(x=target_col, data=df)
plt.title("Meal Type Distribution")
plt.show()

# Pairplot for numerical features
num_cols = ['Calories', 'Protein', 'Fat', 'Carbs', 'Sugar', 'Fiber', 
            'Sodium', 'Cholesterol', 'Glycemic_Index', 'Water_Content']
sns.pairplot(df[num_cols + [target_col]], hue=target_col)
plt.show()
```
